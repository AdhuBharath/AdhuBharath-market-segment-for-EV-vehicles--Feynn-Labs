```python

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
features = ['Battery Capacity (kWh)', 'Range per Charge (km)', 'Charging Time', 'Price', 
            'Power (HP or kW)', 'Top Speed (km/h)', 'Year of Manufacture']
X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []  # within-cluster sum of squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Let's assume the optimal number of clusters is 3 (you can change this based on the Elbow plot)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Example of a 2D visualization using PCA for dimensionality reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Cluster'], palette='viridis')
plt.title('KMeans Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

##### Conclusion:
##### The KMeans clustering analysis on the Indian electric vehicle dataset identified three distinct clusters. Cluster 1 primarily includes vehicles with high battery capacity, range, and power, catering to premium segments. Cluster 2 consists of mid-range vehicles, balancing price, range, and performance, ideal for mainstream consumers. Cluster 3 encompasses budget-friendly options with lower battery capacity and power, suitable for cost-conscious buyers. The PCA visualization confirms these segments, highlighting clear distinctions in vehicle characteristics. This segmentation provides valuable insights for manufacturers to tailor their product offerings and marketing strategies to specific consumer needs in the Indian electric vehicle market.
