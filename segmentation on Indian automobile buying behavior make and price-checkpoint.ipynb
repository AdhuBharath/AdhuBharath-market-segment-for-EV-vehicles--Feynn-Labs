```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans

# Select relevant features for clustering
features = ['Make', 'Price']
X = df1[features]
```

```python
# One-hot encode the 'Make' variable
onehot_encoder = OneHotEncoder()
make_encoded = onehot_encoder.fit_transform(X[['Make']]).toarray()
```

```python
# Combine encoded 'Make' with 'Price'
X_encoded = pd.concat([pd.DataFrame(make_encoded, columns=onehot_encoder.get_feature_names_out(['Make'])), 
                       X[['Price']].reset_index(drop=True)], axis=1)
```

```python
# Normalize the 'Price' data
scaler = StandardScaler()
X_encoded['Price'] = scaler.fit_transform(X_encoded[['Price']])
```

```python
# Determine the optimal number of clusters using the Elbow method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_encoded)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means clustering with the optimal number of clusters 
optimal_clusters = 2 
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(X_encoded)
```

```python
# Add the cluster labels to the original data
df1['Cluster'] = clusters
```

```python
# Analyze and visualize the clusters
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df1, x='Make', y='Price', hue='Cluster', palette='viridis')
plt.title('Clusters of Buyers Based on Make and Price')
plt.xlabel('Make')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Summary statistics for each cluster
cluster_summary = df1.groupby('Cluster')[['Make', 'Price']].agg(lambda x: x.value_counts().index[0])
cluster_summary['Mean Price'] = df1.groupby('Cluster')['Price'].mean()
print(cluster_summary)
```

Based on the clustering analysis using car make and vehicle price as features, we identified two distinct clusters in the automobile buying behavior dataset.

Cluster 0:

Make: Predominantly includes SUVs.
Price: Average price around 1,600,000.
Mean Price: Approximately 1,592,683.
Insights: This cluster represents buyers who prefer SUVs, indicating a preference for larger and potentially more luxurious or versatile vehicles. The higher average price suggests that these buyers have a higher budget and are willing to spend more on their vehicle.

Cluster 1:

Make: Predominantly includes Baleno.
Price: Average price around 700,000.
Mean Price: Approximately 912,241.
Insights: This cluster includes buyers who prefer the Baleno, a more budget-friendly and economical option compared to SUVs. The lower average price indicates that these buyers are likely more price-sensitive and are looking for cost-effective options.

Conclusion:
The clustering analysis reveals two distinct market segments based on car make and vehicle price. Cluster 0 comprises buyers who prefer more expensive SUVs, suggesting a market segment with higher purchasing power and a preference for premium vehicles. Cluster 1, on the other hand, includes buyers who prefer the more affordable Baleno, indicating a market segment that prioritizes cost-effectiveness and practicality. Automobile companies can leverage these insights to tailor their marketing strategies and product offerings, focusing on premium features and luxury for Cluster 0, while emphasizing affordability and efficiency for Cluster 1.
