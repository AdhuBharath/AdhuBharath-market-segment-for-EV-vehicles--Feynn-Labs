```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
```

```python
# Select relevant features for clustering
features = ['Age', 'No of Dependents', 'Total Salary', 'Price']
X1 = df1[features]

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X1)
```

```python
# Determine the optimal number of clusters using the Elbow method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
    
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means clustering with the optimal number of clusters (e.g., 3)
optimal_clusters = 3 
kmeanss = KMeans(n_clusters=optimal_clusters, random_state=42)
clusterss = kmeans.fit_predict(X_scaled)
```

```python
# Add the cluster labels to the original data
df1['Cluster2'] = clusterss
```

```python
# Analyze and visualize the clusters
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df1, x='Age', y='Total Salary', hue='Cluster2', palette='viridis')
plt.title('Clusters of Buyers')
plt.xlabel('Age')
plt.ylabel('Total Salary')
plt.show()
```
